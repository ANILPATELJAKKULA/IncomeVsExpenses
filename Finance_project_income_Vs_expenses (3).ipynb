# This pie chart shows how a group of people are divided by their professions.

```python
plt.figure(figsize=(6,6))
Profession=df["Profession"].value_counts()
plt.pie(Profession.values,labels=Profession.index,autopct="%1.1f%%")
plt.title(" Profession distribution")
plt.legend(loc="lower left", bbox_to_anchor=(1,1))
```

- This is a pie chart showing the percentage distribution of people by profession in your dataset.

- Each slice of the pie represents a different profession, and its size shows the proportion.
  



# **Scatter plot of Applicant Income vs Home Loan EMI**

```python
plt.scatter(df["Annual_Income_(INR)"],df["Home_Loan_EMI_(INR)"])
plt.title("scatter plot of applicant income vs Loan Amount")
plt.xlabel("Annual_Income_(INR)")
plt.ylabel("Home_Loan_EMI_(INR)")
plt.show()
```

# Insights

- Most of the data points are concentrated where the annual income is between ₹4 Lakhs and ₹10 Lakhs.
- most applicants are middle-income earners.
- People with high income may still take small loans.
- Banks might limit EMI based on other factors like credit history or liabilities.

```python
top_investment_cities=df.groupby("City")["Investments_(INR)"].mean().sort_values(ascending=False).head(10)
plt.figure(figsize=(14,8))
sns.barplot(y=top_investment_cities.values, x=top_investment_cities.index,palette="viridis")
plt.title("Top 10 Cities by Average Annual Investment")
plt.xlabel("Average Annual Investment (INR)")
plt.ylabel("City")
plt.show()
```

### → These cities may have better financial literacy, economic growth, or income levels
